두 개의 정수를 받아 합을 반환하는 함수 add를 작성하세요. 매개변수와 반환값에 타입 어노테이션을 추가하세요.

```python
def add(a: int, b: int):
    return a + b


result = add(1 ,2)
result1 = add("1","2")
print(result)
print(result1)
```

def 를 이용해 add()변우를 함수로지정하고 () 안에 a , b 에 각각 : int를 적용해 타입 어노테이션을 적용 했습니다

다시 return 리턴을 하며 a 와 b 를 더하여 결과를 다시 add에 적용합니다

def add(a: int, b: int):
    return a + b

그후 result 결과를 2개를 만들어 가각 (1,2) 와 ("1","2") 를 적용 하여 각각 결과를 하나씩 출력하여 결과를 확인 했습니다

result = add(1 ,2)
result1 = add("1","2")
print(result)
print(result1)
