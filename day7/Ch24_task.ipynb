사용자의 이름, 나이, 성별을 출력하는 함수 print_info를 작성하세요. 이 때, 모든 매개변수는 키워드 매개변수로만 받도록 하세요.

```python
def print_info(**nag):
    return f"이름: {nag['name']}, 나이: {nag['age']}, 성별: {nag['gender']}"

result = print_info(name="김우중", age=31, gender="남자")

print(result)
```
