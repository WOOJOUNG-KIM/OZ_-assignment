사용자로부터 이름과 나이를 입력 받아, "안녕하세요, [이름]님! 당신은 [나이]살입니다."라고 출력하는 함수 greet_user를 작성하세요.

```python
def greet_user():
    name = input("이름을 입력하세요 ")
    age = input("나이를 입력하세요 ")

    print(f"안녕하세요, {name}님! 당신은 {age}살입니다.")

greet_user()
```

def 함수를 정해주는 것을 사용하여 greet_user 변수를 함수로 지정하고

를 만들어야 하니 input 을 이용하여 사용자가 입력할수 있게 코드를 만든다 만들때는 def 코드 안에 해야
