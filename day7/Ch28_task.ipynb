사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
def min_max(a,s,d):
    return min(a,s,d),max(a,s,d)

a= int(input())
s= int(input())
d= int(input())



print(min_max(a,s,d))
```

def 사용해 min_max() 함수로 정해주고

()안에 최솟값 최댓값 을 구할 변수 a,s,d를 넣어

def min_max(a,s,d):

return 리턴을 사용하여 min() 최솟값 max() 최댓값 안에 각각 a,s,d 를 넣어

   return min(a,s,d),max(a,s,d)

각 asd 에 int(input()) 을 이용해 int() 가 정수로 return 이 튜플로 min_max로 보내
