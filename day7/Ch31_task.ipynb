문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_lengths(a):
    for i in a:
        print(len(i))

a = ["코딩","너무나","어렵네요"]
print_lengths(a)
```
