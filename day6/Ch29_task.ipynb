주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
def is_even(num):
    if num % 2 == 0:
        return True
    else:
        return False

result = is_even(2)
print(result)
```
