두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(a1, a2):
    if len(a1) >= len(a2):
        return a1
    else:
        return a2

print(longer_string('연수님화이팅', '인천 연수동 살았었어요'))
print(longer_string('지금은 신제주에서', '코딩공부해요'))
```
